```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv("Exercise_Data.csv")

numeric_df = df.select_dtypes(include="number")

sns.heatmap(numeric_df.corr(), annot=True, cmap="coolwarm")
plt.title("Heatmap of Pulse Data")
plt.show()
```

The heatmap shows how pulse values are related at different times. The values at 15 minutes and 30 minutes are strongly related, which means the heart rate stays similar after exercise. The 1 minute values are less related, showing that the heart rate changes more at the beginning. This helps us understand how heart rate behaves over time.

```python
print(df.columns)
```

```python
df.columns = df.columns.str.strip()

diet_col = [col for col in df.columns if "diet" in col.lower()][0]
pulse_cols = [col for col in df.columns if "min" in col.lower()]

long_df = df.melt(
    id_vars=[diet_col],
    value_vars=pulse_cols,
    var_name="Time",
    value_name="Pulse"
)

# plot
sns.catplot(
    data=long_df,
    x=diet_col,
    y="Pulse",
    col="Time",
    kind="bar",
    height=5,
    aspect=1
)

plt.suptitle("Pulse Values by Diet", y=1.05)
plt.show()
```

The categorical plot shows how diet affects pulse values at different times. The graph makes it easy to compare pulse rates between different diet groups. We can see that diet has an effect on heart rate. This shows that healthy eating habits can impact the body.

```python
planets = sns.load_dataset("planets")
print(planets.head())
```

```python
sns.scatterplot(data=planets, x="distance", y="orbital_period")
plt.title("Distance vs Orbital Period")
plt.show()
```

```python
This scatter plot shows the relationship between distance and orbital period. Planets that are farther away usually take longer time to orbit. This means distance affects how long a planet takes to go around its star.
```

```python
sns.scatterplot(data=planets, x="year", y="mass")
plt.title("Year vs Mass")
plt.show()
```

```python
This graph shows how planet mass changes over different years. It helps us see how discoveries have changed over time. Some years include larger planets while others include smaller ones.
```

```python
sns.histplot(planets["distance"].dropna(), kde=True)
plt.title("Distribution of Distance")
plt.show()
```

```python
This graph shows how distance values are spread. Most planets are closer, and fewer planets are very far away. This means the data is mostly concentrated at smaller distances.
```

```python
sns.histplot(planets["orbital_period"].dropna(), kde=True)
plt.title("Distribution of Orbital Period")
plt.show()
```

```python
This graph shows how orbital period values are spread. Many planets have shorter orbital periods, while fewer have longer ones. This helps us understand how common different orbital periods are.
```

```python
sns.boxplot(data=planets, x="method", y="distance")
plt.xticks(rotation=90)
plt.title("Distance by Method")
plt.show()
```

```python
This boxplot compares planet distance by discovery method. It shows that different methods find planets at different distances. Some methods detect planets farther away than others.
```

```python
sns.boxplot(data=planets, x="method", y="orbital_period")
plt.xticks(rotation=90)
plt.title("Orbital Period by Method")
plt.show()
```

This graph compares orbital period by discovery method. Some methods find planets with shorter periods, while others find planets with longer ones. This shows differences between discovery techniques.
